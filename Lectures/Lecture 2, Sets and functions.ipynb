# Sets
A set is a collection of elements where each element is present without duplicates. This means that a set cannot contain the same element twice. For example, $\{1,2,3,a,b\}$ denotes a set but $\{1,1,2\}$ does not as the element $1$ is duplicated.

Sets can be either finite or infinite. Infinite sets contain an infinite numbers of elements.

In this course we are mainly interested in sets of numbers. The most useful examples are:

### The set of natural numbers
This is the set of numbers you learned to use in primary school. In this course we will consider $0$ to be a natural number.

$$
\mathbb{N} := \{0, 1, 2, 3, 4, ...\}
$$

### The set of integers
Integers are natural numbers with a sign. Each natural number except for $0$ has its negative version:

$$
\mathbb{Z} := \{..., -2, -1, 0, 1, 2, ...\}
$$

We will denote the positive integers as:

$$
\mathbb{Z}^{+} := \{1, 2, 3, 4, ...\}
$$

and the negative integers as:

$$
\mathbb{Z}^{-} := \{-1, -2, -3, ...\}
$$

### The set of rational numbers
Rational numbers are defined as ratios of integers:
$$
\mathbb{Q} := \Big\{\frac{a}{b}\enspace|\enspace a,b\in\mathbb{Z}\wedge b\neq0\Big\} 
$$

We denote the set of rational numbers as $\mathbb{Q}$.

Note that two fractions can represent the same rational number. For example:

$$
\frac{1}{2} = \frac{2}{4}
$$

### The set of real numbers
Not all numbers can be expressed as a ratio of integers. The simplest example is the square root of two: $\sqrt{2}$.

Numbers that are not rational are said to be irrational. The set of real numbers $\mathbb{R}$ contains both rational and irrational numbers. A real number is defined by a (usually infinite) sequence of digits. For example, the square root of two in base $10$ is:

$$
\sqrt{2} = 1.41421356237309504880168...
$$

Note that two sequences of digits can represent the same number, for example:

$$
1.00000000000... = 0.999999999... 
$$

real numbers can be decomposed into a fractional part and a integral part, for example:

$$
\sqrt{2} = 1 + 0.41421356237309504880168...
$$

The true definition of $\mathbb{R}$ is not given since it's too complex for this course.

#### Bonus: Extraordinary proof
More formally, the fractional part of a real number can be defined as the limit of a infinite sum. Consider a infinite sequence of digits:
$$
A = A_1, A_2, A_3, ...
$$
from these digits, we can construct a real number $a$ as the following infinite sum of rational numbers
$$
x = \displaystyle\sum_{n=1}^\infty \frac{A_n}{10^n}
$$
Note that the division by a power of ten places the digits in the right position in the decimal expansion:
$$
\begin{aligned}
&0.\color{blue}{4}\color{red}{1}\color{green}{4}\color{purple}{2}... = \\
&0.\color{blue}{4}000... + \\
&0.0\color{red}{1}00... + \\
&0.00\color{green}{4}0... + \\
&0.000\color{purple}{2}... + \\
\end{aligned}
$$

using this definition, it is easy to see that $1.0000000000...$ = $0.9999999999$:

$$
1.00000... = 1 + \sum_{n=1}^\infty \frac{0}{10^n} = 1
$$

$$
0.99999... = 0 + \sum_{n=1}^\infty \frac{9}{10^n} = 9 \sum_{n=1}^\infty (0.1)^n = 9 \left(\sum_{n=0}^\infty (0.1)^n - 1 \right) = 9 \left(\frac{1}{1 - 0.1} - 1 \right) = 1
$$
where I used the formula for summing the infinite geometric series (see lecture 1).

# Functions

In this lecture we are going to discuss functions. A function between the set $A$ and a set $B$ is defined as a "rule" that assigns one and only one element of $B$ for each element of $A$.

We denote a function from $A$ to $B$ as follows:

$$
f:A \rightarrow B
$$

For each element $a$ of $A$, we denote the corresponding element of $B$ as $f(a)$. The (input) set $A$ is called domain while the (output) set $B$ is called codomain.

The following figure visualizes a function from the set (of numbers) $A = \{m, h, t, a\}$ to the set (of numbers) $B = \{3, 6, 4, 2, 9\}$.

A function can be seen as a (potentially infinite) table of pairs. For example, the function in the image is defined by the following table:
$$
\begin{aligned}
&(m, f(m)) = (m, 3) \\
&(h, f(h)) = (h, 4) \\
&(t, f(t)) = (t, 9) \\
&(a, f(a)) = (a, 2) \\
\end{aligned}
$$

## Injective functions

A function $f:A \rightarrow B$ is said to be injective if, for all different $a_1$ and $a_2$ in $A$, $f(a_1)$ is always different from $f(a_2)$. In other words, a injective function never maps to elements of $A$ on the same element of $B$.

## Surjective functions

The range (or image) of a function is the set of its values. We will denote the range of $f$ as $f(A)$:

$$
f(A) = \{f(a) \enspace | \enspace \forall a\in A\}
$$

Clearly, the range of a function $f: A \rightarrow B$ is a subset of $B$. The function $f$ is said to be surjective when its range is identical to B. This means that all elements of $B$ can be written as $f(a)$ for some $a$ in $A$.

The function in the figure is surjective as $f(A) = \{f(a), f(b)\} = \{1, 2\} = B$.

## Bijective functions and their inverse function

A function is said to be bijective (one-to-one correspondence) when it is both injective and surjective. This means that each element $b$ of $B$ can be written as $f(a)$ in a unique way. 
When a function $f:A \rightarrow B$ is bijective, it is possible to obtain a inverse function $f^{-1}: B \rightarrow A$ such that:

$$
f^{-1}(f(a)) = a
$$

and

$$
f(f^{-1}(b)) = b 
$$

The inverse function is obtained simply by reversing the order of the pairs. For example, the inverse of:

$$
\begin{aligned}
&(1, f(1)) = (1, 1) \\
&(2, f(2)) = (2, 3) \\
&(3, f(3)) = (3, 2)
\end{aligned}
$$

is 

$$
\begin{aligned}
&(1, f^{-1}(1)) = (1, 1) \\
&(3, f^{-1}(3)) = (3, 2) \\
&(2, f^{-1}(2)) = (2, 3)
\end{aligned}
$$

Note that this is possible only if the function is bijective since: 

1) $f^{-1}(b)$ needs to be defined $\forall b\in B$, this is possible only if $f$ is surjective. 

2) $f^{-1}(b)$ has to be uniquely defined. This is only true when $f$ is injective.

### Examples

- The function $f_1(x) = x$ is equal to its own inverse.
- The inverse of $f_2(x) = x + 1$ is $f_2^{-1}(x) = x - 1$.
- The inverse of $f_3(x) = 3 x$ is $f_3^{-1}(x) = \frac{x}{3}$

Constant functions are the simplest possible functions. They map all the input to a single value:

$$

## Linear functions

Linear functions are obtained by multiplying the input by a factor:

$$
f(x) = a x
$$

If the factor $a$ is equal to zero, the function is the $f(x) = 0$ constant function. Otherwise, the linear function is one-to-one and its inverse is given by:

$$
f^{-1}(x) = \dfrac{x}{a}
$$

note that the inverse of a linear function is itself a linear function. Here is the plot of the function $f(x) = 2 x$ and its inverse:

```python
f = lambda x: 2*x
f_inv =  lambda x: x/2

x_range = np.linspace(-1,1,100)
p = 3

plt.plot(x_range, [f(x) for x in x_range], label="f(x) = 2x")
plt.plot(x_range, [f_inv(x) for x in x_range], label="f^-1(x) = x/2")
plt.xlim(min(x_range), max(x_range))
plt.legend(loc="best")
```

## Powers and roots

Powers functions are obtained by taking a integer power of the input:

When the power $p$ is odd, the function is one-to-one and it can be inverted. Conversely, when $p$ is even, the function is one-to-one only if we restrict the domain to non-negative real numbers. The inverse is given by the $p$-th root function:

$$
f_p^{-1}(x) = \sqrt[p]{x}
$$

Note that, when $p$ is even, $\sqrt[p]{x}$ is defined only when $p$ is non-negative.

We can now plot the graph of some power functions:

```python
f = lambda x,p: x**p

for p in range(6):
    plt.plot(x_range, f(x_range, p), label="f(x) = x^{}".format(p))
plt.xlim(min(x_range), max(x_range))
plt.legend(loc="best")
```

and of the corresponding root functions:

```python
def f_inv(x, p):
    if x>= 0:
        return x**(1/p)
    else:
        if p%2 == 0:
            return np.nan
        else:
            return -(abs(x))**(1/p)
        
for p in range(1,5):
    plt.plot(x_range, [f_inv(x, p) for x in x_range], label="f(x) = x^{}".format(p))
plt.xlim(min(x_range), max(x_range))
plt.ylim(-1.5, 1.5)
plt.legend(loc="best")
```

Root functions can also be written as fractional roots:

$$
\sqrt[p]{x} = x^{\frac{1}{p}}
$$

In fact, since roots are the inverse functions of powers, we have that:

$$
(x^p)^{\frac{1}{p}} = x^{\frac{p}{p}} = x^1 = x
$$

## Rational functions

Rational functions are defined as ratios of polynomials:

$$
f(x) = \dfrac{\displaystyle\sum_{p=0}^P \beta_p x^p}{\displaystyle\sum_{k=0}^K \alpha_k x^k}
$$

The domain of a rational function is the set of real number minus a finite number of points corresponding to the roots of the polynomial in the denominator (a root of a function is a point where the value of the function is zero). The domain of a rational function cannot clearly include these points as it would imply a division by zero.

Consider the following rational function:

$$

$$
\mathbb{R}\backslash \{0\}
$$

## The exponential function

The exponential function is positive-valued and its domain is the whole set of real numbers. The most important property of the exponential is that it converts sums into products:

$$
\exp(a+b) = \exp(a)\cdot\exp(b)
$$

From this property, we can see that:

$$
\exp(x) = \exp(0 + x) = \exp(0)\cdot\exp(x)
$$

Hence:

$$
\exp(0) = 1
$$

Furthermore:

$$
1 = \exp(0) = \exp(x + (- x)) = \exp(x)\cdot\exp(-x)
$$

Hence:

$$
\exp(-x) = \dfrac{1}{\exp(x)}
$$


### Exponential function with integer inputs

Let's consider the case where the input of the exponential function is a natural number:

$$
\exp(n)
$$

Using the property of the exponential, we can derive a simple recursive relation:

Thereforem when $n$ is a integer the exponential can be written as the number $\exp(1)$ to the $n$-th power. This number is called the Euler number (denoted as $e$) and it is defined by the following infinite sum:

$$
\mathrm{e} = \exp(1) = \displaystyle\sum_{n=1}^\infty \dfrac{1}{n!}
$$

Therefore, we can write $\exp(n)$ as

$$
\mathrm{e}^n
$$

Similarly, we will sometime write

$$
\exp(x) = \mathrm{e}^x
$$

note that this is just a convenient notation as it does not make sense to say that the number $\mathrm{e}$ should be multiplied by itself $x$ times when $x$ is not a integer.

### The natural logarithm

The natural logarithm is the inverse function of the exponential:

$$
\log(\exp(x)) = \exp(\log(x)) = x
$$
Since the exponential is positive-valued, the domain of the logarithm is the set of positive real numbers. In other words, we cannot define the logarithm of a negative number (or of zero) as the value of the exponential is always positive.

Being the inverse of the exponential, the logarithm converts products into sums:

$$
\log(a \cdot b) = \log(a) + \log(b)
$$

The logarithm also converts ratios into differences:

$$
\log\Big(\dfrac{a}{b}\Big) = \log(a) - \log(b)
$$

There are infinitely many other version of the exponential function all respecting the property:

## Trigonometric functions

The trigonometric functions are strictly connected with circles. The equation for a circle centered at the origin with radius one is:

$$
\color{blue}{x^2} + \color{green}{y^2} = 1
$$

Each point in the circle is uniquely defined by a angle $\color{purple}{\theta}$ ranging from $0$ to $2 \pi$ radians. Points in the circle can then be expressed as function of the angle using the cosine and the sine function:

$$
\color{blue}{\cos(\color{purple}{\theta})^2} + \color{green}{\sin(\color{purple}{\theta})^2} = 1~.
$$


<img src="https://upload.wikimedia.org/wikipedia/commons/thumb/7/72/Sinus_und_Kosinus_am_Einheitskreis_1.svg/500px-Sinus_und_Kosinus_am_Einheitskreis_1.svg.png" alt="drawing" width="300"/>

Therefore, the cosine gives the $x$ coordinate of a point in a circle of radius one while the sine gives the $y$ coordinate.

From the diagram is clear that, when $\color{purple}{\theta}$ is equal to zero, the sine is equal to zero while the cosine is equal to one. Conversely, the cosine is zero and the sine is one when $\color{purple}{\theta}$ is equal to $\pi/2$.

The result are functions that oscillate up and down:

```python
x_range = np.linspace(0,2*np.pi,300)

plt.plot(x_range, np.sin(x_range), lw=2, label="sin(x)", c="g")
plt.plot(x_range, np.cos(x_range), lw=2, label="cos(x)", c="b")
plt.xlim(0,2*np.pi)
plt.legend(loc="best")
plt.title("Trigonometric functions")
```

We can also extend these functions to the whole real axis by assuming that they are periodic with period $2 \pi$.

```python
x_range = np.linspace(-10*np.pi,10*np.pi,300)

plt.plot(x_range, np.sin(x_range), lw=2, label="sin(x)", c="g")
plt.plot(x_range, np.cos(x_range), lw=2, label="cos(x)", c="b")
plt.xlim(-10*np.pi,10*np.pi)
plt.legend(loc="best")
plt.title("Trigonometric functions")
```

### Important trigonometric identities

There are many important identities relating the values of the trigonometric functions. You can find a comprehensive list [here](https://en.wikipedia.org/wiki/List_of_trigonometric_identities).

We already introduced the most important identity:

$$
\cos(x)^2 + \sin(x)^2 = 1 ~.
$$

As you can see from their graphs, sine and cosine can be obtained from each other by shifting the input:

$$
\sin(x) = \cos(x + \pi/2)
$$

$$
\cos(x) = \sin(x - \pi/2)
$$

Other two very important identities can be used for doubling the angle:

$$
\sin(2x) = 2\cos(x)\sin(x)
$$

$$
\cos(2x) = \cos(x)^2 - \sin(x)^2
$$

These are special cases of formulas for summing angles:

$$
\sin(x + y) = \cos(x)\sin(y) + \cos(y)\sin(x)
$$

$$
\cos(x + y) = \cos(x)\cos(y) - \sin(y)\sin(x)
$$

We can also invert the angle doubling formula to get angle halfing formulas:

$$
\sin\Big(\frac{x}{2}\Big)^2 = \frac{1 - \cos(x)}{2}
$$

$$
\cos\Big(\frac{x}{2}\Big)^2 = \frac{1 + \cos(x)}{2}
$$

## The tangent and the inverse trigonometric functions

The tangent is defined as the ratio between sine and cosine:

$$
\tan{(x)} = \dfrac{\sin(x)}{\cos(x)}
$$

Clearly, this expression makes sense only when the denominator is different from zero (you can't divide by zero!). Therefore, the domain of the tangent is the set of real numbers $z$ such that $\cos(z) \neq 0$.

```python
x_range = np.linspace(-2*np.pi,2*np.pi,1000)

f = lambda x: np.tan(x) if np.abs(np.cos(x))>0.01 else np.nan

plt.plot(x_range, [f(x) for x in x_range], lw=2, label="tan(x)", c="g")
plt.xlim(-2*np.pi,2*np.pi)
plt.ylim(-5,5)
plt.legend(loc="best")
plt.title("Tangent")
```

As you can see from the graph, the tangent is periodic with period $\pi$.

### Inverse functions

Sine, cosine and tangents are periodic and therefore they are clearly not one-to-one. This means that we cannot define their inverses. 

For example, there are infinitly many values of $x$ such that:

$$
\cos(x) = 1
$$

In particular, we have that:

$$
\cos(n 2 \pi) = 1
$$

for all possible integers $n$.

# Limits of a function

Limits of functions can be defined in a similar way than limits of sequences. Consider a function $f(x)$ and a point $x_0$. If $l$ is the limit of $f(x)$ at $x_0$, we write:

$$
\displaystyle\lim_{x\rightarrow x_0} f(x) = l
$$ 

## Definition

The function $f(x)$ is said to have limit $l$ for $x$ tending to $x_0$ when:

For all $\epsilon > 0$, there is a $\delta_\epsilon > 0$ such that:

$$
|f(x) - l| \leq \epsilon
$$

for all values of $x$ such that $|x - x_0| < \delta$.

In other words, for any possible small quantity $\epsilon$ there has to be another small quantity $\delta_\epsilon$ such that every time the input $x$ of the function is less than $\delta_\epsilon$ away from the limit point $x_0$, then the value of the function $f(x)$ is less than $\epsilon$ away from the limit $l$.

This can be visualized as follows:

```python
f = lambda x: x**2

x_range = np.linspace(-3,3,100)
plt.plot(x_range, f(x_range), c="k")
a = 1
#plt.scatter(a, f(a), c="r")
epsilon = 0.5
plt.fill_between(x_range, f(a)-epsilon, f(a)+epsilon, alpha = 0.2, color = "b", label = "|f(x) - l| < epsilon")
delta = 0.2
plt.axvspan(a - delta, a + delta, alpha=0.2, color='red', label = "|x - a| < delta")
plt.axvline(x=1, c="r", ls="--", label="x_0 = 1 (Limit point)")
plt.axhline(y=f(1), c="b", ls="--", label="l (Limit)")
plt.xlim(min(x_range), max(x_range))
plt.legend(loc="upper left")
```

The black line is the graph of the function $f(x)$. The red dashed line denotes the location of the limit point $x_0$ while the blue dashed line denotes the value of the limit. The blue shaded area is the region that is not more than $\epsilon$ away from the limit $l$. The definition of the limit says that for any blue shaded areas around the limit there has to be a red shaded area around $x_0$ such that the graph of the function within this red shaded area is fully contained within the blue shaded area.

Let's consider a situation where the limit does not exist and it is not always possible to draw the red shaded area. Consider the following function:

$$
\begin{equation}
  f(x)=\begin{cases}
    x^2, & \text{if x < 1}.\\
    x^2 + 1, & \text{otherwise}.
  \end{cases}
\end{equation}
$$

Let's try to see if this function has a limit for $x$ tending to $x_0 = 1$. We can start with the guess that the limit is equal to $1$ and see what happens. Let's consider $\epsilon = 2$. in this case, we can find a $\delta_\epsilon$ such that the graph is fully contaned in the blue shaded area as far as we are in the red shaded area.

```python
def f(x):
    if x < 1:
        return x**2
    elif x == 1.:
        return np.nan
    else:
        return 1 + x**2

x_range = np.linspace(-3,3,100)
plt.plot(x_range, [f(x) for x in x_range], c="k")
a = 1
l = 1.
plt.scatter(a, l, c="r")
epsilon1 = 2
plt.fill_between(x_range, l-epsilon1, l+epsilon1, alpha = 0.2, color = "b", label = "|f(x) - l| < epsilon_1")
delta1 = 0.2
plt.axvspan(a - delta1, a + delta1, alpha=0.2, color='red', label = "|x - a| < delta_1")
plt.xlim(min(x_range), max(x_range))
plt.axvline(x=1, c="r", ls="--", label="x_0 = 1 (Limit point)")
plt.axhline(y=1, c="b", ls="--", label="l (Tentative limit)")
plt.legend(loc="best")
plt.show()
```

This is a good first step. However, in order to prove that $1$ is really the limit this property needs to hold for all possible values of $\epsilon$. 
Let's see what happens if $\epsilon$ is equal to $0.5$.

```python
plt.plot(x_range, [f(x) for x in x_range], c="k")
a = 1
l = 1.
plt.scatter(a, l, c="r")
epsilon1 = 1.2
delta1 = 0.2
epsilon2 = 0.5
plt.fill_between(x_range, l-epsilon2, l+epsilon2, alpha = 0.2, color = "b", label = "|f(x) - l| < epsilon_2")
plt.xlim(min(x_range), max(x_range))
plt.axvline(x=1, c="r", ls="--", label="x_0 = 1 (Limit point)")
plt.axhline(y=1, c="b", ls="--", label="l (Tentative limit)")
plt.legend(loc="best")
plt.show()
```

In this case you can see that is is impossible to find a $\delta_\epsilon$ that satsfies the property since every small interval around $x_0 = 1$ will always contains a small part on the right side of one that is always more than $1$ away from $l = 1$ since the function has a jump of size one at that point. 

therefore, we can conclude that $1$ is not the limit of this function at $x_0 = 1$.

## Limits at infinity

The limit at infinity is defined in a similar way. A function $f(x)$ has limit $l$ for $x$ tending to $\infty$ when for all $L > 0$, there is a $x_L > 0$ such that:


$$
f(x) > L
$$

for all $x$ bigger than $x_L$. 

This definition can again be visualized using shaded areas:


```python
f = lambda x: x**2/(1 + x**2)

x_range = np.linspace(0,10,100)
plt.plot(x_range, f(x_range), c="k")
l = 1
plt.axhline(l, c = "k", lw = 2, ls = "--", label="Limit")
epsilon = 0.05
plt.fill_between(x_range, l-epsilon, l+epsilon, alpha = 0.2, color = "b", label = "f(x) > L")
X = 5
plt.axvspan(X, max(x_range), alpha=0.2, color='red', label = "x > x_L")
plt.xlim(min(x_range), max(x_range))
plt.legend(loc="best")
```

# Continuous functions

The concept of continuity is very important in calculus. A function $f(x)$ is said to be continuous at a point $x_0$ if the function has a limit for $x$ tending to $x_0$ and the limit is equal to the value of the function.

In formulas, a function $f(x)$ is continuous at $x_0$ when it has a limit at $x_0$ and:
$$
f(x) = \displaystyle\lim_{x \rightarrow x_0} f(x)
$$

The function $f(x)$ is said to be continuous if it is continuous at all possible points. The intuition behind continuous functions is quite straightforward. Continuous are functions never jump!

# Asymptotic relations

We can define asymptotic relations between functions in a similar way we already did in the case of sequences. Let $x_0$ be a point in the domains of the functions $f(x)$ and $g(x)$. $f(x)$ is said to be asymptotic to $g(x)$ for $x$ tending to $x_0$ when:

$$
\displaystyle\lim_{x \rightarrow x_0} \dfrac{f(x)}{g(x)} = \displaystyle\lim_{x \rightarrow x_0} \dfrac{g(x)}{f(x)} = 1
$$

Note that this definition depends on the choice of the limit point $x_0$. For example, $x^2 + x$ is asymptotic to $x$ for $x$ tending to $0$ but it is instead asymptotic to $x^2$ for $x$ tending to infinity:

For $x$ tending to $0$:

$$
\displaystyle\lim_{x \rightarrow 0} \dfrac{x^2 + x}{x} = \displaystyle\lim_{x \rightarrow 0} \left( \dfrac{x^2}{x} + 1 \right) = \left(\displaystyle\lim_{x \rightarrow 0} x \right) + 1 = 0 + 1 = 1
$$

Therefore $x^2 + x$ is asymptotic to $x$ for $x$ tending to $0$.

Conversely, for $x$ tending to infinity:

$$
\displaystyle\lim_{x \rightarrow \infty} \dfrac{x^2 + x}{x^2} = \displaystyle\lim_{x \rightarrow \infty} \left( \dfrac{x^2}{x^2} + \dfrac{1}{x} \right) = 1 + \left(\displaystyle\lim_{x \rightarrow \infty} \dfrac{1}{x} \right) = 1 + 0 = 1
$$

Therefore $x^2 + x$ is asymptotic to $x^2$ for $x$ tending to $\infty$.

In general, when $x$ tends to zero a polynomial is always asymptotic to the monomial with the LOWEST power. Conversely, when $x$ tends to infinity the polynomial is asymptotic to the monomial with the its HIGHEST power.
